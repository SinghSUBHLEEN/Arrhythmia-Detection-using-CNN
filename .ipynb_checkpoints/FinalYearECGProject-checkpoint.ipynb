```python
# Now, we are going to apply discrete time wavelet transform on the baseline corrected signal
# and see how the high frequency and low frequency detailed coefficients turns out to be

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import pywt

with open('/content/rec_2.dat', 'rb') as file:
  data = np.fromfile(file, dtype='int16')

#correcting the baseline of the data

corrected_data = signal.detrend(data)

# creating a butter-worth filter or 2nd order
# these are the best frequencie for this data

cutoff_freq = 15
sample_freq = 500 #Hz

b, a = signal.butter(2, cutoff_freq / (sample_freq / 2))

#filtering the signal

filtered_data = signal.filtfilt(b, a, corrected_data)

[a4, d4, d3, d2, d1] = pywt.wavedec(corrected_data, 'sym4', mode='symmetric', level=4)

# plotting the detailed coefficients

plt.figure(figsize=(14, 3))
plt.plot(corrected_data)
plt.title("Original ECG Signal")
plt.xlabel('samples')
plt.ylabel('amplitude')
plt.grid(True)
plt.tight_layout()

plt.figure(figsize=(14, 3))
plt.plot(a4)
plt.title("Approximated Coefficient 4")
plt.xlabel('samples')
plt.ylabel('amplitude')
plt.grid(True)
plt.tight_layout()

plt.figure(figsize=(14, 3))
plt.plot(d4)
plt.title("Detailed Coefficient 4")
plt.xlabel('samples')
plt.ylabel('amplitude')
plt.grid(True)
plt.tight_layout()

plt.figure(figsize=(14, 3))
plt.plot(d3)
plt.title("Detailed Coefficient 3")
plt.xlabel('samples')
plt.ylabel('amplitude')
plt.grid(True)
plt.tight_layout()

plt.figure(figsize=(14, 3))
plt.plot(d2)
plt.title("Detailed Coefficient 2")
plt.xlabel('samples')
plt.ylabel('amplitude')
plt.grid(True)
plt.tight_layout()

plt.figure(figsize=(14, 3))
plt.plot(d1)
plt.title("Detailed Coefficient 1")
plt.xlabel('samples')
plt.ylabel('amplitude')
plt.grid(True)
plt.tight_layout()

plt.show()

```
