```python
# differentiator and convolutor code 

import numpy as np
import math
import matplotlib.pyplot as plt


def sineWaveGenerator(samples, freq):
    N = samples
    res = np.ones(N)
    
    for i in range(N):
        res[i] = math.sin(2*math.pi*freq*i/N)
        
    return res


def diff(sig):
    diff = np.ones(len(sig)-1)
    
    for i in range(len(sig)-1):
        diff[i] = sig[i+1]-sig[i]
        
    return diff


def convolve(sig1, sig2):
    res = np.zeros(len(sig1)+len(sig2)-1)
    
    # now applying the convolution by taking the point from the signals
    
    for i in range(len(sig1)):
        for j in range(len(sig2)):
            res[i+j] += sig1[i]*sig2[j]
    
    return res


if __name__ == "__main__":

    N = 1000
    freq = 10
    
    signal = sineWaveGenerator(N, freq)
    diff_signal = diff(signal)
    con_signal = convolve(signal, diff_signal)
    
    plt.figure(figsize=(10, 3))
    plt.plot(diff_signal)
    plt.plot(signal)
    plt.tight_layout()
    plt.show()
    
    plt.figure(figsize=(10, 3))
    plt.plot(np.convolve(diff_signal, signal))
    plt.tight_layout()
    plt.show()
```
